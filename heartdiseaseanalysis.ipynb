## Risk Factors for Heart Disease

```python
df2 = pd.DataFrame()
for column in ['Sex', 'AgeCategory', 'SmokerStatus', 'BMI', 'HadHeartAttack', 'HadAngina']:
    df2[column] = df1[column]
    print(df2[column].value_counts())
    print("--------------------------")
    print("--------------------------")
    print("--------------------------")
```

```python
# Handling missing data for columns with missing values
df2['BMI'].fillna(df2['BMI'].mean(), inplace=True)
df2["SmokerStatus"].fillna("Unknown", inplace=True)
df2["AgeCategory"].fillna(df2["AgeCategory"].mode()[0], inplace=True)
df2["HadHeartAttack"].fillna(df2["HadHeartAttack"].mode()[0], inplace=True)
df2["HadAngina"].fillna(df2["HadAngina"].mode()[0], inplace=True)
```

#### Handling Outliers

```python
#Function for converting AgeCategory into a numerical value
def convertAgeCategorytoAge(ageCategory):
    ageCategory_split = ageCategory.split(" ")
    
    # Check if both parts are digits
    if ageCategory_split[1].isdigit() and ageCategory_split[3].isdigit():
        # Calculate the average age in the range
        age_range_start = int(ageCategory_split[1])
        age_range_end = int(ageCategory_split[3])
        return (age_range_start + age_range_end) / 2
    else:
        # Use the starting age if not a range or if the range is '80 or older'
        return int(ageCategory_split[1])
```

```python
df2["AverageAge"] = df2["AgeCategory"].apply(convertAgeCategorytoAge)
```
